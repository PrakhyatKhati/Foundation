```python
import os
from nltk.tokenize import word_tokenize
import string
from collections import Counter
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

```python
### read in the books
books  = {int(f.split(' ')[0]) : f for f in os.listdir('../data/Foundation_201811')}
for i in range(1,8):
    print(i, books[i])
```

```python
### tokenize the books
allwords = []
for i in range(1,8):
    book = books[i]
    for line in open('../data/Foundation_201811/' + book):
        tokens   = word_tokenize(line)
        tokens   = [w.lower() for w in tokens]
        table    = str.maketrans('', '', string.punctuation)
        stripped = [w.translate(table) for w in tokens]
        words    = [word for word in stripped if word.isalpha()]
        words    = [w for w in words if not w in stop_words]        
        if len(words)>0:
            allwords += words
             
print(len(allwords))      
Counter(allwords).most_common(1000)
```

```python
### parse the worlds
worlds = [line.strip().lower() for line in open('../data/worlds_clean.txt')]
worlds
```

```python
### count the worlds
for ind,(a, b) in enumerate(Counter(allwords).most_common(1000)):
    if a in worlds:
        print(ind, a, b)
```
