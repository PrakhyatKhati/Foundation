```python
import os
import pandas as pd
from matplotlib.colors import LinearSegmentedColormap as lsc
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
import numpy as np

def format_axis(ax):   
    for pos in ['right', 'top', 'left']:
        ax.spines[pos].set_edgecolor('white')    
    ax.tick_params(axis='x', length=6, width=2, colors='k')
    ax.tick_params(axis='y', length=6, width=2, colors='k') 
    for tick in ax.xaxis.get_major_ticks():  tick.label.set_fontsize(36) 
    for tick in ax.yaxis.get_major_ticks():  tick.label.set_fontsize(36)
```

```python
### parse the input
books  = {int(f.split(' ')[0]) : f for f in os.listdir('../data/Foundation_201811')}
worlds = [line.strip() for line in open('../data/worlds_clean.txt')]
len(worlds)
```

```python
### number of times each worlds was mentioned in 
### the different sections (at least 20 mentions for the viz)
dfsw = []
names_counts = {}

for i in range(1,8):
    book = books[i]
    with open('../data/Foundation_201811/' + book) as myfile:
        text = myfile.read()
    
    for n in worlds:
        names_counts[n] = text.count(n)

    df = pd.DataFrame([names_counts])
    df.index = [i]
    dfsw.append(df)

dfsw = pd.concat(dfsw)
dfss = dfsw.T
dfss['sum'] = dfss.sum(axis=1)
dfss = dfss[dfss['sum']>19]
tokeep = set(dfss.index)
dfsw = dfsw.T[dfsw.T.index.isin(tokeep)].T
dfsw
```

```python
### visualize the number of mentions of the
### frequently mentioned planets
black  = '#1B1526'
purple = '#D2A7FA'

cmap_cols2 = [black, purple]
cmap2 = lsc.from_list('cmap', cmap_cols2)

dfsw2 = dfsw.copy()
dfsw2 += 1

f, ax = plt.subplots(1,1,figsize=(50,30))
im = ax.imshow(dfsw2,  cmap = cmap2, norm=LogNorm())

names2 = list(dfsw2.keys())
ax.set_xticks(range(len(names2)))
ax.set_xticklabels([a for a in names2], fontsize = 25, rotation = 90, ha = 'center')
ax.set_ylim(-0.5,6.5)
ax.set_yticks([a for a in list(np.arange(0,7))])
ax.set_yticklabels([books[a+1].split('_')[0]  for a in np.arange(0,7)], fontsize = 30, ha = 'left')

for tick in ax.get_xaxis().get_major_ticks():  tick.set_pad(20)
for tick in ax.get_yaxis().get_major_ticks():  tick.set_pad(470)
format_axis(ax)

cbar = f.colorbar(im, ax=ax, shrink=0.42)
for t in cbar.ax.get_yticklabels():
     t.set_fontsize(30)
        
    
#plt.savefig('figures/2_worlds.png', bbox_inches = 'tight', dpi = 100)
```
