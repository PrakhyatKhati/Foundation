```python
import pandas as pd
import os
import numpy as np
import plotly.graph_objects as go
from matplotlib.colors import LogNorm
import gzip
import math
import plotly.express as px
import matplotlib.pyplot as plt

# a basic function to format ax
def format_axis(ax):   
    for pos in ['right', 'top', 'left']:
        ax.spines[pos].set_edgecolor('white')    
    ax.tick_params(axis='x', length=6, width=2, colors='black')
    ax.tick_params(axis='y', length=6, width=2, colors='black') 
    for tick in ax.xaxis.get_major_ticks():  tick.label.set_fontsize(33) 
    for tick in ax.yaxis.get_major_ticks():  tick.label.set_fontsize(33)
```

```python
### parse the cleaned career trajectory based on GoodReads
df    = pd.read_csv('../data/goodreads_cleaned.csv',    index_col = 0)
df_f  = pd.read_csv('../data/goodreads_foundation.csv', index_col = 0)
df_l  = pd.read_csv('../data/goodreads_labeled.csv',    index_col = 0)
df_fl = pd.read_csv('../data/goodreads_foundation_labeled.csv',    index_col = 0)
print(len(df))
df.head(5)
```

```python
### visualize the career history as a stemplot with matplotlib

def stemplot(name, ax, df, marker, color, ecolor, label, alpha = 0.7, markersize = 30, max0 = 0):

    year = list(df.year)
    cnt  = list(df.ratings)
    maxi = max(cnt)
    maxi = 1.1*max([maxi, max0])
    summ = sum(cnt)
    ymin = 15000

    [ax.axvline(x = year[i], ymin = 0, ymax = (cnt[i]+ymin)/(maxi + ymin), alpha = 0.7, linewidth = 3, color = 'grey')  for i in range(len(year))]
    ax.plot(year, cnt, marker, color = color, markersize = markersize, alpha = alpha, markeredgecolor=ecolor,markeredgewidth = 2, label = label )
    ax.set_ylim([-ymin,maxi])
    format_axis(ax)
    
    ax.set_xlabel('Publication year', fontsize = 50, labelpad = 35)
    ax.set_ylabel('Number of ratings', fontsize = 50, labelpad = 45)
    ax.set_yticks([100000, 200000, 300000, 400000, 500000])
    ax.set_yticklabels(['100k', '200k', '300k', '400k', '500k'])
    ax.set_title('Isaac Asimov\'s trajectory', y = 1.3, loc = 'left', fontsize = 60)
    ax.legend(loc = 'best', frameon = False, fontsize = 30)

  
f, ax = plt.subplots(1,1,figsize=(40,11))

grey   = '#2E2618'
purple = '#D2A7FA'
black  = '#1B1526'

stemplot('alma', ax, df,   'o', black, grey, 'All the books with at least 100 ratings', max0 = 0, alpha = 0.9)
stemplot('alma', ax, df_f, 'o', purple,  grey, 'Foundation books', max0 = 0, alpha = 0.85, markersize = 33)
stemplot('alma', ax, df_f, 'o', purple,  grey, label = '', max0 = 0, alpha = 0.6, markersize = 35)

# plt.savefig(folderout + '/1_stemplot_a.png', dpi = 100, bbox_inches = 'tight')
```

```python
### build an interactive version of the stemplot with plotly

def draw_interactive_stemplot(df_l, df_f):

    fig = go.Figure()

    fig.add_trace(go.Scatter(
        x=df_l['year'], y=df_l['ratings'], text=df_l['text'],
        marker_size=9, opacity = 0.76, marker_color=black, name = 'Year:'  ))

    fig.add_trace(go.Scatter(
        x=df_f['year'], y=df_f['ratings'], text=df_f['text'],
        marker_size=9, opacity = 0.6, marker_color=purple,
        name = 'Year',textfont=dict( size=16 )))


    fig.update_traces(mode='markers', marker=dict( size=10,line=dict( color=grey, width=2 ),),)
    fig.update_layout(
        title='Isaac Asimov\'s trajectory',
        xaxis=dict( title='Publication year',  gridcolor='white', gridwidth=2,),
        yaxis=dict( title='Number of ratings', gridcolor='white', type='log', gridwidth=2, ),
        paper_bgcolor='White',
        plot_bgcolor='White',
        width=1000,
        height=400,
        font=dict(  size=16,
        )
    )

    fig.update_layout(hovermode="x unified")
    fig.update_layout(hovermode="y unified")
    fig.show()
    #fig.write_html("figures/1_stemplot.html")


    
draw_interactive_stemplot(df_l, df_fl)
```
